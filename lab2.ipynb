```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from PIL import Image        
```

1) Wyświetlanie obrazu wczytanego z pliku o podanej nazwie,

```python
name = 'cat.jpg'
image_color = Image.open(name)
img_color = np.asarray(image_color)
plt.title('Wczytany obraz')
plt.imshow(img_color)
```

2) Sporzadzenie wykresów zmian poziomu szarosci wzdłuz wybranej linii poziomej lub pionowej o zadanej współrzednej

```python
grayscale = Image.open(name).convert("L")    #convert("L") - konwersja zdjęcia do skali szarości
gray = np.asarray(grayscale)

plt.figure(figsize=(15,10))

plt.subplot(1,2,1) 
plt.title('Obraz')                            
plt.imshow(gray, cmap='gray', vmin=0, vmax=255) #cmap - mapa kolorów, vmin i vmax - zakres, który obejmuje mapa kolorów.

plt.subplot(1,2,2)
plt.title('Histogram')
plt.hist(gray) 
plt.show() 
```

#### Ćwiczenie 7.

Wypróbuj działanie wyrównywania histogramu na przykładowych
obrazach. By zaobserwować skuteczność procedury, poddaj wyrównywaniu obrazy
zbyt ciemne i zbyt jasne. Narysować histogramy obrazów przed i po wyrównaniu.

```python
name = 'dog.jpg'
image = Image.open(name).convert("L")
image_dark = np.asarray(image)

plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Zbyt ciemny obraz')
plt.imshow(image_dark, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title('Histogram')
plt.hist(image_dark)
plt.show()
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Zbyt ciemny obraz po korekcie')
image_correction = cv.equalizeHist(image_dark)                #wyrównywanie 
plt.imshow(image_correction, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title('Histogram')
plt.hist(image_correction)
plt.show()
```

```python
name = 'dog1.jpg'
image = Image.open(name).convert("L")
image_bright = np.asarray(image)

plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Zbyt jasny obraz')
plt.imshow(image_bright, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title('Histogram')
plt.hist(image_bright)
plt.show()
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Zbyt jasny obraz po korekcie')
image_correction = cv.equalizeHist(image_bright)                #wyrównywanie 
plt.imshow(image_correction, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title('Histogram')
plt.hist(image_correction)
plt.show()
```

#### Ćwiczenie 8
a) Zbadaj skuteczność redukcji szumu typu „sól i pieprz” za pomocą liniowego filtra uśredniającego z kwadratową maską

```python
name = 'saltpepper.png'
image = Image.open(name).convert("L")
img = np.asarray(image)

plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Obraz oryginalny')
plt.imshow(img, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title('Maska 3x3')
plt.imshow(cv.blur(img,(3,3)), cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,3)
plt.title('Maska 5x5')
plt.imshow(cv.blur(img,(5,5)), cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,4)
plt.title('Maska 11x11')
plt.imshow(cv.blur(img,(11,11)), cmap='gray', vmin=0, vmax=255)
```

b) Zbadaj skuteczność redukcji szumu typu „sól i pieprz” za pomocą nieliniowego filtra medianowego

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Obraz oryginalny')
plt.imshow(img, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title('Maska 3x3')
plt.imshow(cv.medianBlur(img,3), cmap='gray', vmin=0, vmax=255) # cv.medianBlur() - maska 3x3

plt.subplot(2,2,3)
plt.title('Maska 5x5')
plt.imshow(cv.medianBlur(img,5), cmap='gray', vmin=0, vmax=255) # cv.medianBlur() - maska 7x7

plt.subplot(2,2,4)
plt.title('Maska 11x11')
plt.imshow(cv.medianBlur(img,11), cmap='gray', vmin=0, vmax=255) # cv.medianBlur() - maska 15X15
```

#### Ćwiczenie 9.
a) Filtr Sobela

```python
name = 'fish.jpg'
image = Image.open(name).convert("L")
fish = np.asarray(image)

plt.figure(figsize=(17,12))

plt.subplot(1,3,1)
plt.title('Obraz oryginalny')
plt.imshow(fish, cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,2)
plt.title('Sobel poziomy')
plt.imshow(cv.Sobel(fish, -1, 1, 0, ksize = 3), cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,3)
plt.title('Sobel pionowy')
plt.imshow(cv.Sobel(fish, -1, 0, 1, ksize = 3), cmap='gray', vmin=0, vmax=255)
```

b) Laplasjan

```python
plt.figure(figsize=(17,12))

plt.subplot(1,3,1)
plt.title('Obraz oryginalny')
plt.imshow(fish, cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,2)
plt.title('Maska 5x5')
plt.imshow(cv.Laplacian(fish, -1, ksize = 5), cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,3)
plt.title('Maska 9x9')
plt.imshow(cv.Laplacian(fish, -1, ksize = 9), cmap='gray', vmin=0, vmax=255)
```

**Ćwiczenie 7.**

Wypróbuj działanie wyrównywania histogramu na przykładowych
obrazach. By zaobserwować skuteczność procedury, poddaj wyrównywaniu obrazy
zbyt ciemne i zbyt jasne. Narysować histogramy obrazów przed i po wyrównaniu.

```python
name = 'avatar_dark.jpg'
image = Image.open(name).convert("L")
image_dark = np.asarray(image)

plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Zbyt ciemny obraz')
plt.imshow(image_dark, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title('Histogram')
plt.hist(image_dark)
plt.show()
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Zbyt ciemny obraz po korekcie')
image_correction = cv.equalizeHist(image_dark)                
plt.imshow(image_correction, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title('Histogram')
plt.hist(image_correction)
plt.show()
```

```python
name = 'avatar_bright.jpg'
image = Image.open(name).convert("L")
image_bright = np.asarray(image)

plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Zbyt jasny obraz')
plt.imshow(image_bright, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title('Histogram')
plt.hist(image_bright)
plt.show()
```

**Ćwiczenie 8**

a) Zbadaj skuteczność redukcji szumu typu „sól i pieprz” za pomocą liniowego filtra uśredniającego z kwadratową maską

```python
name = 'avatar_noise.jpg'
image = Image.open(name).convert("L")
img = np.asarray(image)

plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Obraz oryginalny')
plt.imshow(img, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title('Maska 3x3')
plt.imshow(cv.blur(img,(3,3)), cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,3)
plt.title('Maska 5x5')
plt.imshow(cv.blur(img,(5,5)), cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,4)
plt.title('Maska 11x11')
plt.imshow(cv.blur(img,(11,11)), cmap='gray', vmin=0, vmax=255)
```

b) Zbadaj skuteczność redukcji szumu typu „sól i pieprz” za pomocą nieliniowego filtra medianowego

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Obraz oryginalny')
plt.imshow(img, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title('Maska 3x3')
plt.imshow(cv.medianBlur(img,3), cmap='gray', vmin=0, vmax=255) 

plt.subplot(2,2,3)
plt.title('Maska 5x5')
plt.imshow(cv.medianBlur(img,5), cmap='gray', vmin=0, vmax=255) 

plt.subplot(2,2,4)
plt.title('Maska 11x11')
plt.imshow(cv.medianBlur(img,11), cmap='gray', vmin=0, vmax=255)
```

**Ćwiczenie 9.**

a) Filtr Sobela

```python
name = 'avatar.png'
image = Image.open(name).convert("L")
fish = np.asarray(image)

plt.figure(figsize=(17,12))

plt.subplot(1,3,1)
plt.title('Obraz oryginalny')
plt.imshow(fish, cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,2)
plt.title('Sobel poziomy')
plt.imshow(cv.Sobel(fish, -1, 1, 0, ksize = 3), cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,3)
plt.title('Sobel pionowy')
plt.imshow(cv.Sobel(fish, -1, 0, 1, ksize = 3), cmap='gray', vmin=0, vmax=255)
```

b) Laplasjan

```python
plt.figure(figsize=(17,12))

plt.subplot(1,3,1)
plt.title('Obraz oryginalny')
plt.imshow(fish, cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,2)
plt.title('Maska 3x3')
plt.imshow(cv.Laplacian(fish, -1, ksize = 3), cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,3)
plt.title('Maska 9x9')
plt.imshow(cv.Laplacian(fish, -1, ksize = 9), cmap='gray', vmin=0, vmax=255)
```
